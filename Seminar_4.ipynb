```python
#Задача 1. Реализуйте функцию, которая на вход получает матрицу и имя файла, 
#далее записывает данную матрицу в файл и ничего не возвращает.

def fileMatrix(matrix, filename):
    with open(filename, 'w') as fl:
        text = fl.write('\n'.join([str(' '.join([str(el) for el in i])) for i in matrix]))
        fl.close
    return 
matrix = [[1, 2], [3, 4]]
fileMatrix(matrix, 'matrix.txt')

# проверка
with open('matrix.txt') as fl:
    text = fl.read()
print(text)
```

```python
#Задача 4 Реализуйте функцию, которая на вход принимает имя файла, далее выделите множество слов с помощью регулярных выражений 
#и возвращает словарь, где ключ - слово, а значение - число его вхождений в файл.
#Проверить реализацию метода можно на файле story.txt. Распечатайте 10 наиболее частотных слов.
def freqWordsList(filename):
        with open(filename) as fl:
            text_lines = fl.readlines()
        newList = []
        for line in text_lines:
            newList.append(line.split(' '))
        for raw in newList:
            if raw == ['\n']:
                newList.remove(raw)
        wordsSet = []
        import re
        for raw in newList:
            for word in raw:
                word = re.sub(r'\W', '', word) #нужно исключить слова типа "баба-яга", но я пока не понимаю, как
                wordsSet.append(word.lower())
        for el in wordsSet:
            if el == '':
                wordsSet.remove(el)
        wordsList = (sorted(wordsSet))
        from collections import Counter
        frequencyDict = Counter(wordsList)
        return frequencyDict
freqDict = dict(freqWordsList('story.txt'))
sortValues = sorted(list(freqDict.values()))
mostFreqWords = []
for i in range(6):
    for item in freqDict.items():
        if item[1] == sortValues[-1]:
            mostFreqWords.append(item)
            sortValues.remove(sortValues[-1])
tenFreqWords = mostFreqWords[0:10]
print(tenFreqWords)
            

```

```python
#Задача 5. Написать функцию, которая определяет является строка email-ом. 
#Применить данную функцию последовательно к строкам из списка ниже и распечатать строку и ответ функции для неё.
def email(A):
    import re
    emails = []
    for el in A:
        emails.append(re.findall(r'\w+@[a-zA-Z_]+?\.[a-zA-Z]{2,6}', el))
    for el in emails:
        if el == []:
            emails.remove(el) #почему-то удаляет не все пустые вложенные списки
    return emails
strings = ['mail@mail.ru', 'bug@@@com.ru', 'bug@@@com.ru', '@val.ru', 'valid@megapochta.com', 'Just Text2',
           'val@val', 'aa@aa.info', 'bug@@@com.ru', 'val@val.a.a.a.a', '12323123@111[]][] ']
print(strings, email(strings))
```
